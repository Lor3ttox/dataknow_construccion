```python
# Verificación de la existencia de archivos en el volumen
try:
    archivos_volumen_construccion = dbutils.fs.ls(volumen_proceso["volumen"])
    if len(archivos_volumen_construccion) > 0:
        print("Archivos existentes en el volumen de construcción.")
        for archivo in archivos_volumen_construccion:
            print(f"-> {archivo.name}")
    else:
        print("No hay archivos en el volumen de construcción.")
except Exception as e:
    print(f"Error: {e}")
```

```python
# Carga de todos los archivos del directorio de volumenes
df_raw_construccion_completo = (
    spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", ",")
    .option("encoding", "Latin1")
    .load(volumen_proceso["volumen"])
)
df_raw_construccion_X = (
    spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", ",")
    .option("encoding", "Latin1")
    .load(volumen_proceso["volumen"] + "/X.csv")
)
df_raw_construccion_Y = (
    spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", ",")
    .option("encoding", "Latin1")
    .load(volumen_proceso["volumen"] + "/Y.csv")
)
df_raw_construccion_Z = (
    spark.read.format("csv")
    .option("header", "true")
    .option("delimiter", ",")
    .option("encoding", "Latin1")
    .load(volumen_proceso["volumen"] + "/Z.csv")
)
```
